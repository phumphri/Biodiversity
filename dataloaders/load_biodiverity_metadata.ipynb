```python
import psycopg2
import pandas as pd
import numpy as np
import math
import os
import socket
```

```python
sql = """
    insert into belly_button.biodiversity_metadata (    
        sampleid,
        row_num,
        event,
        ethnicity,
        gender,
        age,
        wfreq,
        bbtype,
        location,
        country012,
        zip012,
        country1319,
        impsurace013,
        npp013,
        mmaxtemp013 ,
        pfc013,
        impsurface1319,
        npp1319,
        mmaxtemp1319,
        pfc1319)
    values(
        %s,
        %s,
        %s,
        %s,
        %s,
        %s,
        %s,
        %s,
        %s,
        %s,
        %s,
        %s,
        %s,
        %s,
        %s ,
        %s,
        %s,
        %s,
        %s,
        %s)
    ON CONFLICT DO NOTHING ;
"""
```

```python

# DESKTOP-S08TN4O
hostname = socket.gethostname()
print("socket.hostname():", hostname)
try:
    if (hostname == 'XPS'):   
        conn = psycopg2.connect(os.environ['LOCAL_POSTGRES'])
        print('Connection okay.')
    elif (hostname == 'DESKTOP-S08TN4O'):   
        conn = psycopg2.connect(os.environ['LOCAL_POSTGRES'])
        print('Connection okay.')
    else:
        conn = psycopg2.connect(os.environ['AWS_POSTGRES'])
        print('Connection okay.')
except Exception as e:
    print('Connection failed:', e)
```

```python
try:
    cur = conn.cursor()
    print('Cursor okay.')
except Exception as e:
    print('Cursor failed:', e)
```

```python
try:
    df = pd.read_csv('instructions/Datasets/Belly_Button_Biodiversity_Metadata.csv')
    print('Read csv okay.')
except Exception as e:
    print('Read csv failed:', e)
df.head()
```

```python
row_num = 0
for index, row in df.iterrows():
    
    sampleid = 'BB_' + str(row['SAMPLEID'])
    
    row_num = row_num + 1
    
    event = row['EVENT']
    
    ethnicity = row ['ETHNICITY']
    if (pd.isnull(ethnicity)):
        ethnicity = None
    
    
    gender = row['GENDER']
    if (pd.isnull(gender)):
        gender = None
    
    age = row['AGE']
    if (math.isnan(age)):
        age = None
    else:
        age = int(age)
    
    wfreq = row['WFREQ']
    if (math.isnan(wfreq)):
        wfreq = None
    else:
        wfreq = int(wfreq)
        
    bbtype = row['BBTYPE']
    if (pd.isnull(bbtype)):
        bbtype = None
    
    
    location = row['LOCATION']
    if (pd.isnull(location)):
        location = None
    
    
    country1319 = row['COUNTRY1319']
    if (pd.isnull(country1319)):
        country1319 = None
    
    country012 = row['COUNTRY012']
    if (pd.isnull(country012)):
        country012 = None
    
    zip012 = row['ZIP012']
    
    if (type(zip012) == float):
        if (math.isnan(zip012)):
            zip012 = None
        else:
            zip012 = str(zip012)
            zip012 = zip012.rjust(5, '0')
    
    zip1319 = row['ZIP1319']
    
    if (type(zip1319) == float):
        if (math.isnan(zip1319)):
            zip1319 = None
        else:
            zip1319 = str(zip1319)
            zip1319 = zip1319.rjust(5, '0')
    
    dog = row['DOG']
    if (pd.isnull(dog)):
        dog = None
    
    cat = row['CAT']
    if (pd.isnull(cat)):
        cat = None
    
    impsurface013 = row['IMPSURFACE013']
    if (math.isnan(impsurface013)):
        impsurface013 = None
    else:
        impsurface013 = int(impsurface013)
    
    npp013 = row['NPP013']
    if (pd.isnull(npp013)):
        npp013 = None
    
    mmaxtemp013 = row['MMAXTEMP013']
    if (math.isnan(mmaxtemp013)):
        mmaxtemp013 = None
    
    mmaxtemp1319 = row['MMAXTEMP1319']
    if (math.isnan(mmaxtemp1319)):
        mmaxtemp1319 = None
    
    pfc013 = row['PFC013']
    if (math.isnan(pfc013)):
        pfc013 = None
    
    impsurface1319 = row['IMPSURFACE1319']
    
    if (type(impsurface1319) == str):
        impsurface1319 = None
    elif (math.isnan(impsurface1319)):
        impsurface1319 = None
    else:
        impsurface1319 = int(impsurface1319)
    
    npp1319 = row['NPP1319']    
    if (type(npp1319) == str):
        npp1319 = None
    elif (math.isnan(npp1319)):
        npp1319 = None
    else:
        npp1319 = int(npp1319)
    
    pfc1319 = row['PFC1319']    
    if (type(pfc1319) == str):
        pfc1319 = None
    elif (math.isnan(pfc1319)):
        pfc1319 = None
        
#     print(row_num, age, wfreq, impsurface013)

    try:
        cur.execute(sql, (sampleid,
            row_num,
            event,
            ethnicity,
            gender,
            age,
            wfreq,
            bbtype,
            location,
            country012,
            zip012,
            country1319,
            impsurface013,
            npp013,
            mmaxtemp013 ,
            pfc013,
            impsurface1319,
            npp1319,
            mmaxtemp1319,
            pfc1319))
    except Exception as e:
        print("SQL failed:", e)
        
print('SQL okay.')
```

```python
try:
    conn.commit()
    print('Commit okay.')
except Exception as e:
    print('Commit failed:', e)
```

```python
try:
    conn.close()
    print('Close okay.')
except Exception as e:
    print('Close failed:', e)
```
